```python
# Import libraries
from parrot import Parrot
import torch
import warnings
warnings.filterwarnings("ignore")
```

```python
! pip install git+https://github.com/PrithivirajDamodaran/Parrot.git
```

```python
# For reproducibility
def random_state(seed):
  torch.manual_seed(seed)
  if torch.cuda.is_available():
    torch.cuda.manual_seed_all(seed)

random_state(1234)
```

```python
#Init models (make sure you init ONLY once if you integrate this to your code)
parrot = Parrot(model_tag="prithivida/parrot_paraphraser_on_T5", use_gpu=False)
```

```python
phrases = ["The National Testing Agency (NTA) will conduct the National Eligibility cum Entrance Test (NEET) exam on July 17, 2022. The last date for candidates to submit the NEET application form and pay the application fee for the national-level medical entrance test was May 20, 2022. The application form correction window is now closed, although the National Testing Agency opened a one time window for candidates to make corrections to their 'Category'."]
```

```python
for phrase in phrases:
  print("-"*100)
  print("Input_phrase: ", phrase)
  print("-"*100)
  para_phrases = parrot.augment(input_phrase=phrase)
  for para_phrase in para_phrases:
   print(para_phrase)
```
